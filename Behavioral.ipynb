# Dimensionality reduction

```python
avgs=ndimdf.groupby('item')[appraisals].mean()#ndimdf.groupby('emo')[appraisals].mean()
dimlabels,dimmatrix,axislabels=avgs.columns, avgs.values.T, avgs.index
```

```python
#use affinity propogation to identify clusters in the dimensions
%autoreload
dimclusters, cluster2realmapping=ndimf.affinitypropcluster(dimmatrix, dimlabels, axislabels, dim1=0, dim2=1)
print cluster2realmapping
```

```python
#kmeans
%autoreload
dimk=ndimf.optimalKclusters1(dimmatrix)
if dimk<1:
    dimk=1
clusterdict, dimclusterresults=ndimf.kmeansclustering(dimmatrix, dimlabels, axislabels, numclust=dimk)
print clusterdict
```

```python
#hierarchical clustering
%autoreload
hierarchy=ndimf.hierarchicalcluster(dimmatrix, dimlabels, similarity='euclidean', colorthresh='default')
```

```python
%autoreload
#PCA
pcaobsscores, pcadimloadings, pcarecoveredinput=ndimf.reduction(dimmatrix.T, reductiontype='PCA')
pcaobsscores, pcadimloadings, pcarecoveredinput=ndimf.reduction(dimmatrix.T, ncomp=8, reductiontype='PCA')#set this based on first screen plot
ndimf.plotinputincomponentspace(pcaobsscores, axislabels, item2emomapping)
```

```python
#ica
%autoreload
icaobsscores, icadimloadings, icarecoveredinput=ndimf.reduction(dimmatrix.T, reductiontype='ICA')
icaobsscores, icadimloadings, icarecoveredinput=ndimf.reduction(dimmatrix.T, ncomp=8, reductiontype='ICA')#set this based on first screen plot
ndimf.plotinputincomponentspace(icaobsscores, axislabels, item2emomapping)
```
