# Modeling

```python
allresults={} #save all results to this dict
confmatsaves={}
inputs={'labelcol':'emo', 'iterations':2, 'folds':{'train':[0], 'test':[1], 'nfolds':2}, 'comparisons':NDE_output} #alternative: comparisons-->NDE_output
modeltypes={'37dim': [ndimdf,appraisals],'37dimASD': [ndimasddf,appraisals],'basicemo': [ndimbedf,appraisalsubsets['be']],'valencearousal': [ndimvadf,appraisalsubsets['va']],'valence': [ndimvdf,appraisalsubsets['v']],'arousal': [ndimadf,appraisalsubsets['a']]}
modeltypes['cohmetrix']=[cohmetrixdf,[col for col in cohmetrixdf.columns if col not in fixedcols]]
#modeltypes['naivebayes']=[nbdf, [col for col in nbdf.columns if col not in fixedcols]]
modeltypes['dlsentiment']=[sentimentdf, [col for col in sentimentdf.columns if col not in fixedcols]]
modeltypes['intensity']=[intensitydf, [col for col in intensitydf.columns if col not in fixedcols]]
```

```python
%autoreload
for chosenmodel in [m for m in modeltypes.keys() if m !='37dimASD']:
    #create df that averages responses for individual items
    modeldata,features=modeltypes[chosenmodel][0],modeltypes[chosenmodel][1]
    #### model without generalization
    gencol=None
    allresults=ndimf.classificationwrapper(modeldata,allresults, chosenmodel, inputs['labelcol'], features, gencol, inputs['iterations'], orderedemos, inputs['folds'], rootdir, comparisons=inputs['comparisons'], visualize=ndimvisualize)
    #### model with generalization across items
    gencol='item'
    allresults=ndimf.classificationwrapper(modeldata,allresults, chosenmodel, inputs['labelcol'], features, gencol, inputs['iterations'], orderedemos, inputs['folds'], rootdir, comparisons=inputs['comparisons'], visualize=ndimvisualize)
    key='%s_%s' %(chosenmodel, gencol)
    confmatsaves[chosenmodel]=ndimf.createsimmtxfromconfmat(allresults[key].confmat, allresults[key].orderedlabels)
ndimf.quicksave(confmatsaves, os.path.join(rootdir,'similarityspaces','NDIMsvmerrors.pkl'))
```
