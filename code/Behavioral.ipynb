# Modeling

```python
allresults={} #save all results to this dict
cfg={'labelcol':'emo', 'iterations':10, 'folds':{'train':[0], 'test':[1], 'nfolds':2}, 'comparisons':None} #alternative: comparisons-->NDE_output
modeltypes={'37dim': [ndimdf,appraisals],'37dimASD': [ndimasddf,appraisals],'basicemo': [ndimbedf,appraisalsubsets['be']],'valencearousal': [ndimvadf,appraisalsubsets['va']],'valence': [ndimvdf,appraisalsubsets['v']],'arousal': [ndimadf,appraisalsubsets['a']]}
```

```python
%autoreload
for chosenmodel in [m for m in modeltypes.keys() if m !='37dimASD']:
    #create df that averages responses for individual items
    modeldata=modeltypes[chosenmodel][0].groupby('item').mean()
    modeldata.insert(0, 'item', modeldata.index)
    modeldata.insert(1, 'emo', [item2emomapping[i] for i in modeldata.index.values])
    features=modeltypes[chosenmodel][1]
    #### model without generalization
    gencol=None
    print "\n*******classifying without generalization: %s***********" %(chosenmodel)
    allresults=ndimf.classifiationwrapper(modeldata,allresults, chosenmodel, cfg['labelcol'], features, gencol, cfg['iterations'], orderedemos, cfg['folds'], rootdir, comparisons=cfg['comparisons'], visualize=ndimvisualize)
    #### model with generalization across items
    gencol='item'
    print "\n*******classifying with generalization across item: %s***********" %(chosenmodel)
    allresults=ndimf.classifiationwrapper(modeldata,allresults, chosenmodel, cfg['labelcol'], features, gencol, cfg['iterations'], orderedemos, cfg['folds'], rootdir, comparisons=cfg['comparisons'], visualize=ndimvisualize)
```

# Dimensionality reduction

```python
avgs=ndimdf.groupby('item')[appraisals].mean()#ndimdf.groupby('emo')[appraisals].mean()
dimlabels,dimmatrix,axislabels=avgs.columns, avgs.values.T, avgs.index
```

```python
#use affinity propogation to identify clusters in the dimensions
%autoreload
dimclusters, cluster2realmapping=ndimf.affinitypropcluster(dimmatrix, dimlabels, axislabels, dim1=0, dim2=1)
print cluster2realmapping
```

```python
#kmeans
%autoreload
dimk=ndimf.optimalKclusters1(dimmatrix)
if dimk<1:
    dimk=1
clusterdict, dimclusterresults=ndimf.kmeansclustering(dimmatrix, dimlabels, axislabels, numclust=dimk)
print clusterdict
```

```python
#hierarchical clustering
%autoreload
hierarchy=ndimf.hierarchicalcluster(dimmatrix, dimlabels, similarity='euclidean', colorthresh='default')
```

```python
%autoreload
#PCA
pcaobsscores, pcadimloadings, pcarecoveredinput=ndimf.reduction(dimmatrix.T, reductiontype='PCA')
pcaobsscores, pcadimloadings, pcarecoveredinput=ndimf.reduction(dimmatrix.T, ncomp=8, reductiontype='PCA')#set this based on first screen plot
ndimf.plotinputincomponentspace(pcaobsscores, axislabels, item2emomapping)
```

```python
#ica
%autoreload
icaobsscores, icadimloadings, icarecoveredinput=ndimf.reduction(dimmatrix.T, reductiontype='ICA')
icaobsscores, icadimloadings, icarecoveredinput=ndimf.reduction(dimmatrix.T, ncomp=8, reductiontype='ICA')#set this based on first screen plot
ndimf.plotinputincomponentspace(icaobsscores, axislabels, item2emomapping)
```
