```python
import sys
sys.path.append('/mindhive/saxelab/scripts/aesscripts/')
import FGE_MISC.code.behave.NDIM as ndimf
import FGE_MISC.code.behave.NDE as ndef
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import FGE_MISC.code.vizfuncs as viz
from FGE_MISC.code.config import vizcfg
```

```python
#helpful functions
def defineitem2emomapping(stimdf):
    item2emomapping={}
    for index,row in stimdf.iterrows():
        item2emomapping[row['item']]=row['answer']
    return item2emomapping
```

# Load Results

```python
rootdir='/mindhive/saxelab/scripts/aesscripts/FGE_MISC/'
resultsdir=os.path.join(rootdir, 'results')
simdir=os.path.join(rootdir, 'similarityspaces')
NDE_output=ndimf.quickload(os.path.join(resultsdir,'NDE_output.pkl'))
orderedemos=list(NDE_output['csummary']['item'].index.values)
item2emomapping=defineitem2emomapping(NDE_output['summary'])
allresults=ndimf.quickload(os.path.join(resultsdir,'NDIMRESULTS.pkl'))
inputspaces=ndimf.quickload(os.path.join(simdir, 'RSAmatrixinfo.pkl'))
allrsasimspaces=ndimf.quickload(os.path.join(simdir, 'comparisonRDMs.pkl'))
explicitdict=ndimf.quickload(os.path.join(simdir, 'NDIMexplicits.pkl'))
```

# Visualize Results

```python
%autoreload
#item wise NDE results
viz.plotconfusions(NDE_output, orderedemos)
plt.tight_layout()
viz.plotaccuracy(NDE_output)
# collapsed emo-wise NDE results
viz.plotcollapsedresults(NDE_output, orderedemos)
plt.tight_layout()
```

```python
%autoreload
#visualize input feature spaces
ax1=viz.visualizeinputs(inputspaces, orderedemos, item2emomapping, ncols=4, collapse='emo')
```
