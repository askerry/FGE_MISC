```python
import NDIM as ndimf
import NDE as ndef
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#helpful functions
def defineitem2emomapping(stimdf):
    item2emomapping={}
    for index,row in stimdf.iterrows():
        item2emomapping[row['item']]=row['answer']
    return item2emomapping
```

# Load Results

```python
rootdir='/Users/amyskerry/dropbox/FGE_MISC/'
resultsdir=os.path.join(rootdir, 'results')
simdir=os.path.join(rootdir, 'similarityspaces')
NDE_output=ndimf.quickload(os.path.join(resultsdir,'NDE_output.pkl'))
orderedemos=list(NDE_output['csummary']['item'].index.values)
item2emomapping=defineitem2emomapping(NDE_output['summary'])
allresults=ndimf.quickload(os.path.join(resultsdir,'NDIMRESULTS.pkl'))
inputspaces=ndimf.quickload(os.path.join(simdir, 'RSAmatrixinfo.pkl'))
allrsasimspaces=ndimf.quickload(os.path.join(simdir, 'comparisonRDMs.pkl'))
explicitdict=ndimf.quickload(os.path.join(simdir, 'NDIMexplicits.pkl'))
```

# Visualize Results

```python
%autoreload
#item wise NDE results
ndef.plotconfusions(NDE_output, orderedemos)
ndef.plotaccuracy(NDE_output)
# collapsed emo-wise NDE results
ndef.plotcollapsedresults(NDE_output, orderedemos)
```

```python
%autoreload
#visualize input feature spaces
ax1=ndimf.visualizeinputs(inputspaces, orderedemos, item2emomapping, ncols=3, collapse='emo')
```

```python
%autoreload
#visualize model classification results
for key in allresults.keys():
    result=allresults[key]
    ndimf.visualizeclassification(result.confmat, result.labelaccs, orderedemos, comparisons=NDE_output)
    plt.tight_layout()
    plt.show()
```

```python
%autoreload
#visualize similarity spaces
ax3=ndimf.visualizeRDMs(allrsasimspaces, orderedemos, ncols=3)   
```

```python
#compare models to eachother
%autoreload
ndimf.plotacccomparison(allresults, flags=['None', 'item'], benchmark=NDE_output['csummary']['accuracy'].mean())
```
