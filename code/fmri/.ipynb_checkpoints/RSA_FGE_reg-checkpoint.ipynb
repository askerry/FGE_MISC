```python
### in progress
rankcolors=False
def bootstrapRDM(grouprdms, iterations, subjlist, roi):
    iterrdms=[]
    if configspec['neuraltype']=='rawsim':
        individual_roi_rmds=grouprdms[roi]
    else:
        raise RuntimeError('implemented for neuraltype==rawsim only')
    subjs=samplesizes=range(len(individual_roi_rmds))
    for i in range(iterations):
        sample=np.array(np.random.choice(subjs, len(grouprdms), replace=True))
        subset_roi_rdms=np.array(individual_roi_rmds)[sample]
        iterrdms.append(np.mean(subset_roi_rdms,0))
    return iterrdms
def computeacrossiterationvariability(matrices, observedmatrix):
    diffmaps=[]
    for iterationmatrix in matrices: #for each iteration
        diffmaps.append(np.abs(observedmatrix-iterationmatrix)) #compute cell-wise different between the iteration and the "true" observed
    diffmaps=np.array(diffmaps)
    varmap=np.mean(diffmaps, axis=0) #we have a single map of differences per iteraction, and we are going to average them
    return varmap
def computewithinmatrixvariability(observedmatrix):
    dim=len(observedmatrix)
    observedmatrix=np.array(observedmatrix).flatten()
    withindiffmap=observedmatrix*0
    for eln,el in enumerate(observedmatrix): # for each cell in the observed matrix
        diffs=np.abs(observedmatrix-el) #compute the difference between it and all other cells in the matrix
        diffs[eln]=np.nan # we don't want to inlude the difference with itself (which is zero)
        withindiffmap[eln]=np.nanmean(diffs) #for that cell, we take the mean of it's differences with every other cell, and we create a matrix of these means
    withindiffmap=withindiffmap.reshape(dim,dim)
    return withindiffmap
def withinminusacrossdiffs(varmap, withindiffmap):
    withindiffmap=withindiffmap.flatten()
    varmap=varmap.flatten()
    f, ax=plt.subplots(2,1, figsize=[10,4], sharex=True)
    ax[1].hist(withindiffmap,bins=100)
    ax[1].set_title('avg across cell differences (in observed matrix)')
    ax[0].hist(varmap,bins=100)
    ax[0].set_title('avg across iteration (within cell) differences')
    ratios=np.divide(withindiffmap.flatten(),varmap.flatten()) 
    ratiosmap=ratios.reshape(20,20)
    #diffs=withindiffmap.flatten()-varmap.flatten() #this should be positive if the differences within the matrix are larger than the dfferences across iterations
    matrixmean=np.mean(ratios)
    f, ax=plt.subplots(figsize=[10,2])
    matrixsem=np.std(ratios,ddof=1)/np.sqrt(len(ratios))
    ax.set_title('distribution of within matrix diffs/across iteration diffs (for each cell)')
    ax.hist(ratios,bins=100)
    print "mean ratio: %.3f 1/- %.2f SEM (across cells)" %(matrixmean, matrixsem)
    return ratiosmap
```

```python
roi='DMPFC_tomloc'
niterations=100
obsmat=np.mean(grouprdms[roi],0)
obsmats=bootstrapRDM(grouprdms, niterations, subjlist, roi)
```

```python
varmap=computeacrossiterationvariability(obsmats, obsmat)
withindiffmap=computewithinmatrixvariability(obsmat)
rmap=withinminusacrossdiffs(varmap, withindiffmap)
```
