```python
import NDIM as ndimf
import NDE as ndef
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import vizfuncs as viz
```

```python
#helpful functions
def defineitem2emomapping(stimdf):
    item2emomapping={}
    for index,row in stimdf.iterrows():
        item2emomapping[row['item']]=row['answer']
    return item2emomapping
```

# Load Results

```python
rootdir='/Users/amyskerry/dropbox/FGE_MISC/'
resultsdir=os.path.join(rootdir, 'results')
simdir=os.path.join(rootdir, 'similarityspaces')
NDE_output=ndimf.quickload(os.path.join(resultsdir,'NDE_output.pkl'))
orderedemos=list(NDE_output['csummary']['item'].index.values)
item2emomapping=defineitem2emomapping(NDE_output['summary'])
allresults=ndimf.quickload(os.path.join(resultsdir,'NDIMRESULTS.pkl'))
inputspaces=ndimf.quickload(os.path.join(simdir, 'RSAmatrixinfo.pkl'))
allrsasimspaces=ndimf.quickload(os.path.join(simdir, 'comparisonRDMs.pkl'))
explicitdict=ndimf.quickload(os.path.join(simdir, 'NDIMexplicits.pkl'))
```
